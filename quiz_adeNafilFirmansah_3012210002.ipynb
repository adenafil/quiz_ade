```python
import matplotlib.pyplot as plt
import seaborn as sns

#set menjadi whitegrid style pada seaborn
sns.set(style="whitegrid")
df_anak_penerima_usia_dini = df['penerima_kriteria_anak_usia_dini']
```

# Visualisasi 1: Distribusi penerima PKH berdasarkan kriteria 'Anak Usia Dini'


```python
# atur figsize 10/6
plt.figure(figsize=(10, 6))
# create visualisasi histplot
sns.histplot(df_anak_penerima_usia_dini, bins=15, kde=True, color='skyblue')
# set judul
plt.title('Distribusi Penerima PKH Berdasarkan Kriteria Anak Usia Dini')
# set x
plt.xlabel('penerima_kriteria_anak_usia_dini')
# set y
plt.ylabel('Frekuensi')
# show
plt.show()
```

# Visualisasi 2: Boxplot penerima PKH berdasarkan kriteria 'Anak Usia Dini'


```python
# create visualisasi boxplot dengan warna pink
# set x dengan df di awal
sns.boxplot(x=df_anak_penerima_usia_dini, color='pink')
# set title
plt.title('Boxplot Penerima PKH Berdasarkan Kriteria Anak Usia Dini')
# set x
plt.xlabel('Jumlah Penerima Anak Usia Dini')
# show
plt.show()
```

# Visualisasi 3: Scatter plot penerima PKH anak usia dini dan ibu hamil (asumsi kolom 'Ibu Hamil' ada)


```python
# create scatter plot dengan df yang ada bulan
sns.scatterplot(x=df_yang_ada_bulan['penerima_kriteria_anak_usia_dini'], y=df_yang_ada_bulan['penerima_kriteria_ibu_hamil'], data=df, hue=df_yang_ada_bulan['bulan'], palette='coolwarm')
# set title
plt.title('Scatter Plot Penerima PKH: Anak Usia Dini vs. penerima_kriteria_ibu_hamil')
# set x
plt.xlabel('Jumlah Penerima Anak Usia Dini')
# set y
plt.ylabel('penerima_kriteria_ibu_hamil')
# show
plt.show()
```

# Visualisasi 4: Bar Plot Jumlah penerima kriteria anak usia dini based on bulan


```python
# buat barplot dengan data df yang ada bulan
sns.barplot(x=df_yang_ada_bulan['bulan'], y=df_yang_ada_bulan['penerima_kriteria_anak_usia_dini'], data=df, ci=None, palette='viridis')
# set title
plt.title('Jumlah penerima kriteria anak usia dini based on bulan')
# set x
plt.xlabel('Bulan')
# set y
plt.ylabel('anak usia dini')
# show
plt.show()
```

# Visualisasi 5: Pie Chart dengan beberapa filtering categories

```python
# set filter index
categories = ['penerima_kriteria_anak_usia_dini	', 'penerima_kriteria_pendidikan_sd', 'penerima_kriteria_pendidikan_smp', 'penerima_kriteria_pendidikan_sma', 'penerima_kriteria_lanjut_usia']
# lakukan perulangan dengan index column tertentu dan sudah di sum
sums = [totak_penerima_kriteria_anak_usia_dini for cat in categories]

# create pie cart
plt.pie(sums, labels=categories, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
# set title
plt.title('Proporsi Penerima Berdasarkan Kategori')
# set axis
plt.axis('equal')
#show
plt.show()
```

#4. Gunakan data berikut : https://raw.githubusercontent.com/PacktPublishing/hands-on-exploratory-data-analysis-with-python/master/Chapter%207/titanic.csv

```python
# Mengunduh dan membaca data ke dalam DataFrame
url_no_4 = "https://raw.githubusercontent.com/PacktPublishing/hands-on-exploratory-data-analysis-with-python/master/Chapter%207/titanic.csv"
titanic = pd.read_csv(url_no_4)

titanic
```

#a. Hitung jumlah missing data.

```python
missing_data = titanic.isnull().sum()
missing_data
```

#b. Hitung persentasi perempuan yang survive.

```python
# cari wanita yang selamat
women = titanic.loc[titanic.Sex == 'female']["Survived"]
# bagi hasil total wanita yang selamat dengan panjang
rate_women = sum(women)/len(women)

rate_women
```

#c. Hitung persentase laki laki yang survive.

#percentage of men survived


```python
# cari pria yang selamat
men = titanic.loc[titanic.Sex == 'male']["Survived"]
# bagi hasil total pria yang selamat dengan panjang
rate_men = sum(men)/len(men)

rate_men
```

d. Visualisasikan jumlah penumpang yang survive dan yang meninggal dunia dari Pclass
yang berbeda.
# gagal

```python
titanic['Survived'] = titanic['Survived'].map({0: "not_survived", 1: "survived"})

fig, ax = plt.subplots(1, 2, figsize=(12, 6))

# Plot jumlah penumpang berdasarkan jenis kelamin
sns.barplot(x=titanic["Sex"].value_counts().index, y=titanic["Sex"].value_counts().values, ax=ax[0], palette='coolwarm')
ax[0].set_title("Number Of Passengers By Sex")
ax[0].set_ylabel("Population")

# Plot jumlah penumpang selamat dan tidak selamat berdasarkan jenis kelamin
sns.countplot(x="Sex", hue="Survived", data=titanic, palette='coolwarm', ax=ax[1])
ax[1].set_title("Sex: Survived vs Dead")
ax[1].set_ylabel("Population")

plt.show()
```

#e. Lakukan analisis multivariat pada variabel Survived, Pclass,Fear and Age.

```python
sns.set(style="ticks", color_codes=True)
sns.pairplot(titanic,height=2,vars = [ 'Fare','Age','Pclass'], hue="Survived")
plt.show()
```
